```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
# Đảm bảo các đường dẫn này đúng với cấu trúc project trên Kaggle
import config
from utils import load_trained_model # Giả định hàm này load model từ file .h5 hoặc .keras
from data_operations.data_preprocessing import load_inbreast_data_no_pectoral_removal

# =================================================================
# BƯỚC 1: CẤU HÌNH VÀ TẢI LẠI MÔ HÌNH ĐÃ HUẤN LUYỆN
# =================================================================
print("--- 1. Tải lại mô hình đã huấn luyện ---")

# --- Đặt các thông số để tải đúng model ---
# Tên của lần chạy tốt nhất
# LƯU Ý: Hãy thay thế bằng tên chính xác của file model bạn đã lưu
# Ví dụ: 'INbreast_ROI_SGD_BalancedReg' hoặc 'INbreast_ROI_SGD_BalancedReg.keras'
MODEL_NAME = 'INbreast_ROI_SuperAug' 
MODEL_PATH = f'/kaggle/input/superaug/tensorflow2/default/1/dataset-INbreast_type-all_model-MobileNet_lr-0.0005_b-16_e1-40_e2-80_roi-True_INbreast_ROI_SuperAug.weights.h5'
mock_args.cli_args = mock_args # Cần cho lỗi 'cli_args is not defined'

# Tải lại model
# Lưu ý: Cần có file kiến trúc json nếu bạn lưu riêng
try:
    # Nếu bạn có file kiến trúc json:
    # with open(f'/path/to/{MODEL_NAME}_arch.json', 'r') as json_file:
    #     loaded_model_json = json_file.read()
    # model = tf.keras.models.model_from_json(loaded_model_json)
    # model.load_weights(MODEL_PATH)
    
    # Nếu bạn có thể load trực tiếp (ví dụ từ CnnModel class)
    # Đây là cách đơn giản hơn nếu CnnModel có thể khởi tạo lại kiến trúc
    from cnn_models.cnn_model import CnnModel
    cnn = CnnModel("MobileNet", num_classes=2, args=mock_args)
    cnn.model.load_weights(MODEL_PATH)
    model = cnn.model
    model.summary()
    print(f"Tải thành công model từ: {MODEL_PATH}")

except Exception as e:
    print(f"Lỗi khi tải model: {e}")
    # Dừng ở đây nếu không tải được model
    model = None

# =================================================================
# BƯỚC 2: TẢI VÀ CHUẨN BỊ DỮ LIỆU TEST
# =================================================================
if model:
    print("\n--- 2. Chuẩn bị dữ liệu để visualize ---")
    DATA_DIR = '/kaggle/input/breastdata/INbreast/INbreast'
    le = LabelEncoder()
    
    # Tải lại toàn bộ dữ liệu (chúng ta chỉ cần tập test)
    X_loaded, y_text_labels = load_inbreast_data_no_pectoral_removal(
        data_dir=DATA_DIR,
        label_encoder=le,
        use_roi_patches=True, # Chạy với ROI
        target_size=(224, 224)
    )
    
    le.fit(y_text_labels)
    y_numeric_labels = le.transform(y_text_labels)

    # Chia lại y hệt như lúc train để có tập test chính xác
    _, X_test, _, y_test_numeric = train_test_split(
        X_loaded, y_numeric_labels, test_size=0.2, stratify=y_numeric_labels, random_state=42
    )
    print(f"Đã tải và chia xong, có {len(X_test)} ảnh trong tập test.")

# =================================================================
# BƯỚC 3: ĐỊNH NGHĨA HÀM GRAD-CAM
# =================================================================
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def display_gradcam(img, heatmap, alpha=0.4):
    # Chuyển ảnh về dạng uint8 để hiển thị
    img_u8 = (img * 255).astype(np.uint8)
    
    # Resize heatmap về kích thước ảnh gốc
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    # Trộn heatmap và ảnh gốc
    superimposed_img = cv2.addWeighted(cv2.cvtColor(img_u8, cv2.COLOR_RGB2BGR), 1-alpha, heatmap, alpha, 0)
    
    return cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB)


# =================================================================
# BƯỚC 4: TẠO VÀ HIỂN THỊ VISUALIZATION
# =================================================================
if model and 'X_test' in locals():
    print("\n--- 4. Tạo và hiển thị Grad-CAM ---")
    
    # Tên của lớp conv cuối cùng trong MobileNetV2
    # Bạn có thể tìm tên này bằng cách `model.summary()`
    LAST_CONV_LAYER_NAME = "out_relu" 

    # Chọn một vài ảnh từ tập test để visualize
    # Ví dụ: 2 ảnh Benign và 2 ảnh Malignant
    benign_indices = np.where(y_test_numeric == 0)[0]
    malignant_indices = np.where(y_test_numeric == 1)[0]
    
    sample_indices = np.concatenate([
        np.random.choice(benign_indices, 2, replace=False),
        np.random.choice(malignant_indices, 2, replace=False)
    ])

    plt.figure(figsize=(12, 12))
    
    for i, img_idx in enumerate(sample_indices):
        original_img = X_test[img_idx]
        img_array = np.expand_dims(original_img, axis=0) # Thêm chiều batch

        # Dự đoán để lấy nhãn
        preds = model.predict(img_array)
        predicted_class_idx = np.argmax(preds[0])
        predicted_class_name = le.classes_[predicted_class_idx]
        true_class_name = le.classes_[y_test_numeric[img_idx]]

        # Tạo heatmap
        heatmap = make_gradcam_heatmap(img_array, model, LAST_CONV_LAYER_NAME)
        
        # Overlay heatmap lên ảnh
        superimposed_img = display_gradcam(original_img, heatmap, alpha=0.5)

        # Hiển thị
        plt.subplot(4, 2, 2*i + 1)
        plt.imshow(original_img)
        plt.title(f"Original - True: {true_class_name}")
        plt.axis('off')

        plt.subplot(4, 2, 2*i + 2)
        plt.imshow(superimposed_img)
        plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter

# --- CÁC IMPORT TỪ PROJECT CỦA BẠN ---
# Đảm bảo các đường dẫn này đúng với cấu trúc project trên Kaggle
import config
from utils import load_trained_model # Giả định hàm này load model từ file .h5 hoặc .keras
from data_operations.data_preprocessing import load_inbreast_data_no_pectoral_removal

# =================================================================
# BƯỚC 1: CẤU HÌNH VÀ TẢI LẠI MÔ HÌNH ĐÃ HUẤN LUYỆN
# =================================================================
print("--- 1. Tải lại mô hình đã huấn luyện ---")

# --- Đặt các thông số để tải đúng model ---
# Tên của lần chạy tốt nhất
# LƯU Ý: Hãy thay thế bằng tên chính xác của file model bạn đã lưu
# Ví dụ: 'INbreast_ROI_SGD_BalancedReg' hoặc 'INbreast_ROI_SGD_BalancedReg.keras'
MODEL_NAME = 'INbreast_ROI_SuperAug' 
MODEL_PATH = f'/kaggle/working/BreastMammo/saved_models/dataset-INbreast_type-all_model-MobileNet_lr-0.001_b-16_e1-40_e2-60_roi-True_INbreast_ROI_SGD_HeavyAug.h5'
mock_args.cli_args = mock_args # Cần cho lỗi 'cli_args is not defined'

# Tải lại model
# Lưu ý: Cần có file kiến trúc json nếu bạn lưu riêng
try:
    # Nếu bạn có file kiến trúc json:
    # with open(f'/path/to/{MODEL_NAME}_arch.json', 'r') as json_file:
    #     loaded_model_json = json_file.read()
    # model = tf.keras.models.model_from_json(loaded_model_json)
    # model.load_weights(MODEL_PATH)
    
    # Nếu bạn có thể load trực tiếp (ví dụ từ CnnModel class)
    # Đây là cách đơn giản hơn nếu CnnModel có thể khởi tạo lại kiến trúc
    from cnn_models.cnn_model import CnnModel
    cnn = CnnModel("MobileNet", num_classes=2, args=mock_args)
    cnn.model.load_weights(MODEL_PATH)
    model = cnn.model
    print(f"Tải thành công model từ: {MODEL_PATH}")

except Exception as e:
    print(f"Lỗi khi tải model: {e}")
    # Dừng ở đây nếu không tải được model
    model = None

# =================================================================
# BƯỚC 2: TẢI VÀ CHUẨN BỊ DỮ LIỆU TEST
# =================================================================
if model:
    print("\n--- 2. Chuẩn bị dữ liệu để visualize ---")
    DATA_DIR = '/kaggle/input/breastdata/INbreast/INbreast'
    le = LabelEncoder()
    
    # Tải lại toàn bộ dữ liệu (chúng ta chỉ cần tập test)
    X_loaded, y_text_labels = load_inbreast_data_no_pectoral_removal(
        data_dir=DATA_DIR,
        label_encoder=le,
        use_roi_patches=True, # Chạy với ROI
        target_size=(224, 224)
    )
    
    le.fit(y_text_labels)
    y_numeric_labels = le.transform(y_text_labels)

    # Chia lại y hệt như lúc train để có tập test chính xác
    _, X_test, _, y_test_numeric = train_test_split(
        X_loaded, y_numeric_labels, test_size=0.2, stratify=y_numeric_labels, random_state=42
    )
    print(f"Đã tải và chia xong, có {len(X_test)} ảnh trong tập test.")

# =================================================================
# BƯỚC 3: ĐỊNH NGHĨA HÀM GRAD-CAM
# =================================================================
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def display_gradcam(img, heatmap, alpha=0.4):
    # Chuyển ảnh về dạng uint8 để hiển thị
    img_u8 = (img * 255).astype(np.uint8)
    
    # Resize heatmap về kích thước ảnh gốc
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    # Trộn heatmap và ảnh gốc
    superimposed_img = cv2.addWeighted(cv2.cvtColor(img_u8, cv2.COLOR_RGB2BGR), 1-alpha, heatmap, alpha, 0)
    
    return cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB)


# =================================================================
# BƯỚC 4: TẠO VÀ HIỂN THỊ VISUALIZATION
# =================================================================
if model and 'X_test' in locals():
    print("\n--- 4. Tạo và hiển thị Grad-CAM ---")
    
    # Tên của lớp conv cuối cùng trong MobileNetV2
    # Bạn có thể tìm tên này bằng cách `model.summary()`
    LAST_CONV_LAYER_NAME = "out_relu" 

    # Chọn một vài ảnh từ tập test để visualize
    # Ví dụ: 2 ảnh Benign và 2 ảnh Malignant
    benign_indices = np.where(y_test_numeric == 0)[0]
    malignant_indices = np.where(y_test_numeric == 1)[0]
    
    sample_indices = np.concatenate([
        np.random.choice(benign_indices, 2, replace=False),
        np.random.choice(malignant_indices, 2, replace=False)
    ])

    plt.figure(figsize=(12, 12))
    
    for i, img_idx in enumerate(sample_indices):
        original_img = X_test[img_idx]
        img_array = np.expand_dims(original_img, axis=0) # Thêm chiều batch

        # Dự đoán để lấy nhãn
        preds = model.predict(img_array)
        predicted_class_idx = np.argmax(preds[0])
        predicted_class_name = le.classes_[predicted_class_idx]
        true_class_name = le.classes_[y_test_numeric[img_idx]]

        # Tạo heatmap
        heatmap = make_gradcam_heatmap(img_array, model, LAST_CONV_LAYER_NAME)
        
        # Overlay heatmap lên ảnh
        superimposed_img = display_gradcam(original_img, heatmap, alpha=0.5)

        # Hiển thị
        plt.subplot(4, 2, 2*i + 1)
        plt.imshow(original_img)
        plt.title(f"Original - True: {true_class_name}")
        plt.axis('off')

        plt.subplot(4, 2, 2*i + 2)
        plt.imshow(superimposed_img)
        plt.title(f"Grad-CAM - Predicted: {predicted_class_name}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Nhãn thật và nhãn dự đoán (ví dụ bạn thay bằng dữ liệu thực tế của bạn)
y_true = ['benign'] * 44 + ['malignant'] * 20
y_pred = (
    ['benign'] * 29 + ['malignant'] * 15 +  # true benign: 29 đúng, 15 sai
    ['benign'] * 7 + ['malignant'] * 13     # true malignant: 7 sai, 13 đúng
)

# Các nhãn lớp
labels = ['benign', 'malignant']

# Tính confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=labels)

# Normalize theo hàng
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# Hiển thị theo dạng x/y
support = cm.sum(axis=1)
annot = np.array([[f'{cm[i, j]}/{support[i]}' for j in range(len(labels))] for i in range(len(labels))])

# Vẽ heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm_normalized, annot=annot, fmt='', cmap='Blues',
            xticklabels=labels, yticklabels=labels, cbar=True)

plt.xlabel('Predicted classes')
plt.ylabel('True classes')
plt.title('Normalized MobileNet Confusion Matrix (with ratios)')
plt.tight_layout()
plt.show()
```
